```python
main_folder = '_Organize IDK'
#father_folder = 'Music_2023'

import os
import pandas as pd
import shutil 

def get_all_paths(directory):
    all_paths = []

    for dirpath, dirnames, filenames in os.walk(directory):
        for dirname in dirnames:
            all_paths.append(os.path.join(dirpath, dirname))
        for filename in filenames:
            all_paths.append(os.path.join(dirpath, filename))

    return all_paths

def create_dataset(all_paths):
    # Initialize lists for the dataset columns
    names = []
    file_types = []
    paths = []
    extensions = []
    sizes = []

    for path in all_paths:
        # Extract file name from path
        name = os.path.basename(path)
        
        # Get the first 15 characters of the file name
        short_name = name[:15]
        names.append(short_name)
        
        # Check if it's a directory or file for the 'Type' column
        if os.path.isdir(path):
            file_type = 'Directory'
            ext = ''  # Directories don't have an extension
            size = 0  # Set size to 0 for directories
        else:
            file_type = 'File'
            _, ext = os.path.splitext(name)  # Split the name and its extension
            size = os.path.getsize(path) / (1024 * 1024)  # Convert bytes to MB
        
        file_types.append(file_type)
        paths.append(path)
        extensions.append(ext)
        sizes.append(size)

    # Convert lists to a pandas DataFrame
    df = pd.DataFrame({
        'Name': names, 
        'Type': file_types, 
        'Path': paths, 
        'Extension': extensions, 
        'Size (MB)': sizes
    })
    
    return df

folder_path = f'{desk_path}/3_Yeriko_dj/ALL_MUSIC/Music_back_up/_1_Source_DJ/_Organize IDK/'  # Replace 'main_folder' with your actual folder name
all_paths = get_all_paths(folder_path)

# Create dataset
df = create_dataset(all_paths)

# Sort Data Frame by size
df = df.sort_values(by='Size (MB)', ascending=False)

# Print the dataset and the length of all_paths
print(len(all_paths))
df.head(107)

# list of FILES PATHS to extract STEMS
song_paths = df['Path'].tolist()
song_paths
print(song_paths,'\n\n################# of FILES --> ',len(song_paths))
```
